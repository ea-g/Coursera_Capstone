## Part 1: scraping and cleaning our table <a name="Part1"></a>

Let's use the `read_html` method in pandas to grab our tables from the page. Afterwards, let's get the right table.

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M', header=0)
torontodf= tables[0]
torontodf.columns=['Postalcode', 'Borough', 'Neighborhood']
torontodf.head()
```

### Cleaning the data:

First, let's drop all the rows which **do not** have a borough assigned. We'll reassign "Not assigned" as a np.nan value.
Afterwards, we'll drop the rows that have nan in borough.

```python
torontodf.replace(to_replace='Not assigned', value=np.nan, inplace=True)
torontodf.dropna(subset=['Borough'], inplace=True)
torontodf.reset_index(drop=True, inplace = True)
torontodf.head(10)
```

Good! Now that that's finished, we'll fill the nan values in **Neighborhood** with their corresponding value from **Borough**

```python
torontodf['Neighborhood'] = torontodf['Neighborhood'].fillna(torontodf['Borough'])
torontodf.head(10)
```

Now, we'll group our neighbourhoods together by postal code and combine these rows so that each postal code is only listed once
with a list of its neighbourhoods. After that, we'll clean up the neighbourhoods so that they're only separated by a comma.

Lastly, we'll check that our new dataframe's shape matches the number of unique postal codes in Toronto. 

**This should be 103 postal codes and 103 rows in the new frame.**

```python
df = torontodf.groupby(['Postalcode', 'Borough'])['Neighborhood'].apply(list).to_frame()
df.reset_index(inplace=True)
df['Neighborhood'] = df['Neighborhood'].astype('str').str.replace('[\[\]]', '', regex=True) #remove []
df['Neighborhood'] = df['Neighborhood'].str.replace('"', '', regex=True) #remove the ""
df['Neighborhood'] = df['Neighborhood'].str.replace("\s'*", ' ', regex=True) #remove ' after a comma
df['Neighborhood'] = df['Neighborhood'].str.replace("\A'", '', regex=True) #remove ' at the start of strings
df['Neighborhood'] = df['Neighborhood'].str.replace("'\Z", '', regex=True) #remove ' at the end of strings
df['Neighborhood'] = df['Neighborhood'].str.replace("', ", ', ', regex=True) #remove ' before a comma
print ("Our cleaned and grouped dataframe shape is:", df.shape, "matching the row count with unique postal codes:", len(torontodf['Postalcode'].unique()))

df.head(10)
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
    
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

### Let's get the venues for all of our neighborhoods in the Toronto boroughs dataframe

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
toronto_venues = getNearbyVenues(names=tboroughs['Neighborhood'],
                                   latitudes=tboroughs['Latitude'],
                                   longitudes=tboroughs['Longitude']
                                  )
```

### Frequency of venue type in Neighborhoods

From here we're going to perform one-hot encoding to get the occurence of a venue type by checking each venue in each neighborhood grouping. Afterwards, we'll get the mean frequency of the venue type. 

```python
# one hot encoding
toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe, notice that neighborhood was a venue category so we need a different name
toronto_onehot['Neighborhoods'] = toronto_venues['Neighborhood'] 

# move neighborhood column to the first column
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

```python
toronto_grouped = toronto_onehot.groupby('Neighborhoods').mean().reset_index()
toronto_grouped
```

### Let's go ahead and print each neighborhood along with the top 5 venue types for that neighborhood(s)

```python
num_top_venues = 5

for hood in toronto_grouped['Neighborhoods']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighborhoods'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

### Convert it to a dataframe

Looks like a lot of coffee shops...
We'll sort the venues in descendeing order and display the top 10 types for the neighborhoods.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]

num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighborhoods']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhoods'] = toronto_grouped['Neighborhoods']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

## Clustering the neighborhoods with k-means

We'll run k-means to make 5 clusters, add those cluster labels to the dataframe, and then visualize with those cluster labels on a map.

```python
# set number of clusters
kclusters = 5

toronto_grouped_clustering = toronto_grouped.drop('Neighborhoods', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

### Merging dataframes
Let's merge our dataframes to include the cluster labels, borough, neighborhood, lat, long, and top ten venue types

```python
# add clustering labels
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = tboroughs
toronto_merged['Neighborhoods'] = toronto_merged['Neighborhood']
toronto_merged.drop(columns='Neighborhood', inplace=True)
#merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhoods'), on='Neighborhoods')

toronto_merged.head() # check the last columns!
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhoods'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
